# Actividad Evaluable: Mapas de calor y boxplots

## 1. Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charging data
df = pd.read_csv('TC1002S/covid19_tweets.csv')

# Converting boolean columns to int
df['user_verified'] = df['user_verified'].astype(int)
df['is_retweet'] = df['is_retweet'].astype(int)

# Displaying the first 5 rows
df.head()
```

## 2. Realiza el an√°lisis de las variables usando diagramas de cajas y bigotes, histogramas y mapas de calor.

### Diagramas de cajas y bigotes, as√≠ como histogramas

### user_followers

```python
# Box plot of user_followers
sns.boxplot(data=df['user_followers'], showfliers=False)
```

```python
# Histogram of user_followers
fig, ax = plt.subplots()
ax.hist(df['user_followers'])
plt.show()
```

### user_friends

```python
# Box plot of user_friends
sns.boxplot(data=df['user_friends'], showfliers=False)
```

```python
# Histogram of user_friends
fig, ax = plt.subplots()
ax.hist(df['user_friends'])
plt.show()
```
